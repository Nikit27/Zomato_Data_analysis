```python
rating=data.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating count"})
rating
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,10)
sns.barplot(x='Aggregate rating',y='Rating count',data=rating)
```

```python
count_red=0
count_yel=0
count_gr=0
count_dg=0
count_wh=0
count_or=0

for i in range(len(rating['Rating color'])):
    if rating["Rating color"][i]=='Red':
        count_red=count_red+rating['Rating count'][i]
    elif rating["Rating color"][i]=='Orange':
        count_or=count_or+rating['Rating count'][i]
    elif rating["Rating color"][i]=='Yellow':
        count_yel=count_yel+rating['Rating count'][i]
    elif rating["Rating color"][i]=='Green':
        count_gr=count_gr+rating['Rating count'][i]
    elif rating["Rating color"][i]=='White':
        count_wh=count_wh+rating['Rating count'][i]
    else:
        count_dg=count_dg+rating['Rating count'][i]
d={"Poor":count_red,"Not Rated":count_wh,"Average":count_or,"Good":count_yel,"Very Good":count_gr,"Excellent":count_dg}
rating_count=pd.DataFrame(d,index=["rating count"]).T


rating_count
```

```python
#Observations
#Most of the people rated average
#a big amount of people have not rated 
#Excellent rating is very less

```

```python
key=list(rating_count.index)
value=list(rating_count["rating count"])
sns.barplot(key,value)
```
